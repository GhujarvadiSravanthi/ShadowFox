**Matplotlib**
Matplotlib is a foundational plotting library in Python. It provides a flexible and powerful platform for creating static, animated, and interactive visualizations.
****
**Features:**

* Highly customizable plots.

* Integration with other libraries (NumPy, Pandas).

* Wide range of plot types.
****

Typical Use Cases: **bold text**

* Basic data visualization.

* Scientific plotting.

**âœ… Line Plot**

Description: Displays data points connected by straight lines.
**Use Case:** Trend visualization.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

**âœ… Scatter Plot**

Description: Shows data as a collection of points.

**Use Case:** Correlation analysis.

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

**âœ… Histogram**

Description: Shows frequency distribution of a dataset.

**Use Case:** Distribution analysis.

```python
data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data, bins=4)
plt.title("Histogram")
plt.show()
```

**Pie Chart**

Description: Displays proportions as slices of a pie.

**Use Case:** Percentage distribution.

```python
sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()
```

**Seaborn**

Seaborn is a statistical data visualization library built on top of Matplotlib. It offers a high-level interface for drawing attractive and informative statistical graphics.

**Features:**

* Simplified syntax for complex plots.

* Built-in themes for aesthetics.

* Integration with Pandas DataFrames.

T**ypical Use Cases:**

* Statistical analysis.

* Exploratory data analysis (EDA).

**Line Plot**

Description: Same as Matplotlib but with improved styling.

**Use Case:** Trend visualization with built-in aesthetics

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x=[1, 2, 3], y=[4, 5, 6])
plt.title("Seaborn Line Plot")
plt.show()
```

**Scatter Plot**

Description: Adds support for color grouping and additional dimensions.

**Use Case:** Relationship and distribution.

```python
sns.scatterplot(x=[1, 2, 3], y=[3, 2, 5], hue=["A", "B", "A"])
plt.title("Seaborn Scatter Plot")
plt.show()
```

**Bar Plot**

Description: Combines bar chart with confidence intervals.

**Use Case:** Category comparison with stats.

```python
sns.barplot(x=['A', 'B', 'C'], y=[10, 20, 15])
plt.title("Seaborn Bar Plot")
plt.show()
```

**Histogram / Distplot**

Description: Combines histogram with KDE for smooth curve.

**Use Case:** Distribution analysis.

```python
sns.histplot([1, 2, 2, 3, 3, 3, 4], bins=4, kde=True)
plt.title("Seaborn Histogram")
plt.show()
```

**Box Plot**

Description: Displays distribution using quartiles.

**Use Case:** Outlier detection and spread.

```python
sns.boxplot(y=[1, 2, 2, 3, 4, 5, 6])
plt.title("Box Plot")
plt.show()
```
